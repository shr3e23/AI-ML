```python
def deep_flatten(lst):
    """
    Flatten a deeply nested list into a single list.

    Parameters:
        lst (list): A nested list of arbitrary depth.

    Returns:
        list: A flattened version of the list.
    """
    flat_list = []
    for element in lst:
        if isinstance(element, list):
            flat_list.extend(deep_flatten(element))
        else:
            flat_list.append(element)
    return flat_list


nested = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
result = deep_flatten(nested)
print(result)
```

8. Nested List Average:

Write a Python function that calculates the average of all elements in a nested list.

• Requirements:

– Define a function average nested(lst) that takes a nested list lst and returns the average
of all the elements.

– Example: For the input [[1, 2], [3, 4], [5, 6]], the output should be 3.5.

```python
def average_nested(lst):
    """
    Calculate the average of all elements in a nested list (any depth).

    Parameters:
        lst (list): A nested list containing numbers or other lists.

    Returns:
        float: The average of all numeric elements.
    """
    def helper(sublist):
        total = 0
        count = 0
        for element in sublist:
            if isinstance(element, list):
                sub_total, sub_count = helper(element)
                total += sub_total
                count += sub_count
            else:
                total += element
                count += 1
        return total, count

    total, count = helper(lst)
    return total / count if count != 0 else 0


nested = [[1, 2], [3, 4], [5, 6]]
result = average_nested(nested)
print(result)
```

3. Dot Product

• Using Python Lists, compute the dot product of two lists of size 1, 000, 000. Measure and
Print the time taken for this operation.

• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for
this operation.

```python
import numpy as np
import time

N = 1_000_000

list1 = list(range(N))
list2 = list(range(N))

start_time = time.time()
dot_list = sum(a * b for a, b in zip(list1, list2))
end_time = time.time()

print("Python list dot product:", dot_list)
print("Time taken (Python lists):", end_time - start_time, "seconds")

arr1 = np.arange(N)
arr2 = np.arange(N)

start_time = time.time()
dot_numpy = np.dot(arr1, arr2)
end_time = time.time()

print("\nNumPy array dot product:", dot_numpy)
print("Time taken (NumPy arrays):", end_time - start_time, "seconds")
```

4. Matrix Multiplication

• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Mea-
sure and print the time taken for this operation.

• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000.
Measure and print the time taken for this operation.

```python
import numpy as np
import time

# Matrix size
N = 1000

# Create two matrices using lists
listA = [[i for i in range(N)] for _ in range(N)]
listB = [[j for j in range(N)] for _ in range(N)]

# Matrix multiplication using nested loops
start_time = time.time()
list_result = [[sum(a * b for a, b in zip(rowA, colB))
                for colB in zip(*listB)]
               for rowA in listA]
end_time = time.time()

print("Python list matrix multiplication time:", end_time - start_time, "seconds")

arrA = np.array(listA)
arrB = np.array(listB)

start_time = time.time()
numpy_result = np.dot(arrA, arrB)
end_time = time.time()

print("NumPy matrix multiplication time:", end_time - start_time, "seconds")
```
